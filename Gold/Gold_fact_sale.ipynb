#Business Transformation based on Modeling

```python
query = """
SELECT
    sc.order_number,
    pr.product_key,
    cu.customer_key,
    sc.order_date,
    sc.ship_date,
    sc.due_date,
    sc.sales_amount,
    sc.quantity,
    sc.price
FROM silver.crm_sale sc
LEFT JOIN gold.dim_product pr
    ON sc.product_number = pr.product_number
LEFT JOIN gold.dim_customer cu
    ON sc.customer_id = cu.customer_id;
"""
df = spark.sql(query)
```

## Sanity Checks Dataframe

```python
df.limit(10).display()
```

# Writing Gold Table

```python
df.write.mode("overwrite").format("delta").option("overwriteSchema", "true").saveAsTable("workspace.gold.fact_sale")
```
