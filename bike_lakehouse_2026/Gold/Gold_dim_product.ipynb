#Business Transformation based on Modeling

```python
query = """
SELECT
    ROW_NUMBER() OVER (ORDER BY pr.product_start_date, pr.product_number) AS product_key, 
    pr.product_id,
    pr.product_number,
    pr.product_name,
    pr.category_id,
    pc.category,
    pc.subcategory,
    pc.requires_maintenance,
    pr.product_line,
    pr.product_start_date
FROM silver.crm_product pr
LEFT JOIN silver.erp_product_category pc
    ON pr.category_id = pc.category_id
"""
df = spark.sql(query)
```

## Sanity Checks Dataframe

```python
df.limit(10).display()
```

# Writing Gold Table

```python
df.write.mode("overwrite").format("delta").option("overwriteSchema", "true").saveAsTable("workspace.gold.dim_product")
```
